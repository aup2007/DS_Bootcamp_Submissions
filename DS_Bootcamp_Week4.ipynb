##  Create a line plot of ZN and INDUS in the housing data.

##  Create a bar chart using col1 and col2 of dummy data.

## a. Give the plot a large title of your choosing.

```python
plt.figure(figsize=(10, 6))
plt.bar(categories, data["col1"], label='col1', color='b', alpha=0.7)
plt.bar(categories, data["col2"], label='col2', color='g', alpha=0.7, bottom=data["col1"])
plt.title('Bar Chart For Dummy Data')
plt.legend()

plt.show()
```

## b. Move the legend to the lower-left corner.


```python
plt.figure(figsize=(10, 6))
plt.bar(categories, data["col1"], label='col1', color='b', alpha=0.7)
plt.bar(categories, data["col2"], label='col2', color='g', alpha=0.7, bottom=data["col1"])
plt.title('Bar Chart For Dummy Data')
plt.legend(loc='lower left')

plt.show()
```

## c. Do the same thing but with horizontal bars.


```python
plt.figure(figsize=(10, 6))
plt.barh(categories, data["col1"], label='col1', color='b', alpha=0.7)
plt.barh(categories, data["col2"], label='col2', color='g', alpha=0.7, left=data["col1"])
plt.title('Bar Chart with Horizontal Bars')
plt.legend()


plt.show()
```
